# Method 1 :  
### Split the columns into categorical and numerical

```python
# Split the columns into categorical and numerical
categorical_columns = dataset.select_dtypes(include=['object']).columns
numerical_columns = dataset.select_dtypes(include=['number']).columns

# Display the categorical columns
print(f"Categorical Columns: {categorical_columns}")

# Display the numerical columns
print(f"\nNumerical Columns: {numerical_columns}")

```

# Method 2 :  Using class,functions & logics
### Split the columns into categorical and numerical

```python
quantitative_columns = []
qualitative_columns = []

for column in dataset.columns:
    if dataset[column].dtype == 'object':
        qualitative_columns.append(column)
    else:
        quantitative_columns.append(column)
```

```python
qualitative_columns, quantitative_columns
```

```python
#This logic works perfectly to separate the columns into qualitative and quantitative categories.
# now lets create a function to reuse this logic
def qualitative_quantitative(dataset):
    quantitative = []
    qualitative = []
    
    for column in dataset.columns:
        if dataset[column].dtype == 'object':
            qualitative.append(column)
        else:
            quantitative.append(column)
    
    return qualitative, quantitative
# Now we can use the function to separate columns in our dataset
qualitative_columns, quantitative_columns = qualitative_quantitative(dataset)
# Display the separated columns
print(f"Qualitative Columns: {qualitative_columns}\n")
print(f"Quantitative Columns: {quantitative_columns}")
```
