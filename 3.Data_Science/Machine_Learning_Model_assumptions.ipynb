# Machine Learning Model Assumptions

### Before creating a ML model there are few things to be taken care or noted

##### 1. NULL or NAN values should not be present

##### 2. Outlier should not be present    -   (25,30,45,200)  there should not be a vast differnce between each data points.

##### **An outlier is an observation that lies an abnormal distance from other values in a random sample from a population.**


#### There are two types of outliers : **Highest Range outlier    &     Lowest Range outlier.**

### Most machine learning algorithms do not work well in the presence of outlier.

3. #### Measure of Central Tendency  :


##### Mean | Arithmetic Mean  ->  Average which gives overall central value.

##### Median   ->  Midpoint of data. 

##### Mode   ->  Repitation of the datepoints from a dataset . 


##### **Concidering Median is great to use when the last dataset has a outlier  (Median tends to omit the outlier)**

###### **But when in the same last dataset has a outlier and you use mean then its of helpless (Mean will not ignore the outlier)**

###### ***For example in Salary dataset using Median is always good - because every employee doesnt not have same range of salary***
