# Four QUARTILE - Q1 Q2 Q3 Q4

```python
descriptive_stats = pd.DataFrame(index = ["Mean", "Median", "Mode", "Q1:25%", "Q2:50%", "Q3:75%", "Q4:100%"], columns=quantitative_columns)
for column in quantitative_columns:
    descriptive_stats.loc["Mean", column] = dataset[column].mean()
    descriptive_stats.loc["Median", column] = dataset[column].median()
    descriptive_stats.loc["Mode", column] = dataset[column].mode()[0]
    descriptive_stats.loc["Q1:25%", column] = dataset.describe()[column]["25%"]
    descriptive_stats.loc["Q2:50%", column] = dataset.describe()[column]["50%"]
    descriptive_stats.loc["Q3:75%", column] = dataset.describe()[column]["75%"]
    descriptive_stats.loc["99%", column] = np.percentile(dataset[column], 99)
    descriptive_stats.loc["Q4:100%", column] = dataset.describe()[column]["max"]
descriptive_stats
```

Reason for outliers could be human errors or the dataset behavours itself like that (Lesser or greater)
# Advanced stage of percentile - IQR (INTER QUARTILE RANGE)

### (preprocessing) In order to remove or replacing the outlier be it - (Lesser Outlier or Greater Outlier)






```python
descriptive_stats = pd.DataFrame(index = ["Mean", "Median", "Mode", "Q1:25%", "Q2:50%", "Q3:75%", "Q4:100%", "IQR", "1.5_rule", "Lesser_outlier", "Greater_outlier", "Min", "Max"], columns=quantitative_columns)
for column in quantitative_columns:
    descriptive_stats[column,"Mean"] = dataset[column].mean()
    descriptive_stats[column,"Median"] = dataset[column].median()
    descriptive_stats[column,"Mode"] = dataset[column].mode()[0]
    descriptive_stats[column,"Q1:25%"] = dataset.describe()[column]["25%"]
    descriptive_stats[column,"Q2:50%"] = dataset.describe()[column]["50%"]
    descriptive_stats[column,"Q3:75%"] = dataset.describe()[column]["75%"]
    descriptive_stats[column,"99%"] = np.percentile(dataset[column], 99)
    descriptive_stats[column,"Q4:100%"] = dataset.describe()[column]["max"]
    descriptive_stats[column,"IQR"] = descriptive_stats[column,"Q3:75%"] - descriptive_stats[column,"Q1:25%"]
    descriptive_stats[column,"1.5_rule"] = 1.5 * descriptive_stats[column,"IQR"]
    descriptive_stats["Lesser_outlier", column] = descriptive_stats[column,"Q1:25%"] - descriptive_stats[column,"1.5_rule"]
    descriptive_stats["Greater_outlier", column] = descriptive_stats[column,"Q3:75%"] + descriptive_stats[column,"1.5_rule"]
    descriptive_stats["Min", column] = dataset.describe()[column]["min"]
    descriptive_stats["Max", column] = dataset.describe()[column]["max"]
descriptive_stats
```

```python
## Check and find out the Lesser and Greater outliers

lesser_outliers = []
greater_outliers = []
for column in quantitative_columns:
    if descriptive_stats.loc["Lesser_outlier", column] > dataset[column].min():
        lesser_outliers.append(column)
    if descriptive_stats.loc["Greater_outlier", column] < dataset[column].max():
        greater_outliers.append(column)
print(f"Lesser Outliers: {lesser_outliers}")
print(f"Greater Outliers: {greater_outliers}")
```

```python
# Now replace the outliers using the IQR method

IQR = []
for column in quantitative_columns:
    IQR.append(descriptive_stats.loc["IQR", column])
# Replace Lesser Outliers with Q1 - 1.5 * IQR
for column in lesser_outliers:
    dataset[column] = dataset[column].apply(lambda x: descriptive_stats.loc["Q1:25%", column] - 1.5 * descriptive_stats.loc["IQR", column] if x < descriptive_stats.loc["Lesser_outlier", column] else x)
# Replace Greater Outliers with Q3 + 1.5 * IQR
for column in greater_outliers:
    dataset[column] = dataset[column].apply(lambda x: descriptive_stats.loc["Q3:75%", column] + 1.5 * descriptive_stats.loc["IQR", column] if x > descriptive_stats.loc["Greater_outlier", column] else x)
# Display the updated dataset
print(dataset[quantitative_columns].describe())
```

# FREQUENCY | RELATIVE FREQUENCY | CUMULATIVE FREQUENCY

```python
dataset[column].value_counts()
```

```python
Frequency_table = pd.DataFrame(columns=["unique_values", "Frequency", "Relative_Frequency", "Cumulative_Frequency"])

Frequency_table
```

```python
def frequency_table(column, dataset):    
    Frequency_table = pd.DataFrame(columns=["unique_values", "Frequency", "Relative_Frequency", "Cumulative_Frequency"])
    
    Frequency_table["unique_values"] = dataset[column].value_counts().index
    Frequency_table["Frequency"] = dataset[column].value_counts().values
    Frequency_table["Relative_Frequency"] = (Frequency_table["Frequency"] / 103)
    Frequency_table["Cumulative_Frequency"] = Frequency_table["Relative_Frequency"].cumsum()
    
    return Frequency_table
```

# Histogram, Skewness, Kurtosis








```python
def Univariate(dataset,quantitative_columns):
    descriptive_stats=pd.DataFrame(index=["Mean","Median","Mode","Q1:25%","Q2:50%",
                               "Q3:75%","99%","Q4:100%","IQR","1.5rule","Lesser_outlier","Greater_outlier",
                                    "Min","Max","kurtosis","skew","Var","Std"],columns=quantitative_columns)
    for column in quantitative_columns:
        descriptive_stats[column]["Mean"]=dataset[column].mean()
        descriptive_stats[column]["Median"]=dataset[column].median()
        descriptive_stats[column]["Mode"]=dataset[column].mode()[0]
        descriptive_stats[column]["Q1:25%"]=dataset.describe()[column]["25%"]
        descriptive_stats[column]["Q2:50%"]=dataset.describe()[column]["50%"]
        descriptive_stats[column]["Q3:75%"]=dataset.describe()[column]["75%"]
        descriptive_stats[column]["99%"]=np.percentile(dataset[column],99)
        descriptive_stats[column]["Q4:100%"]=dataset.describe()[column]["max"]
        descriptive_stats[column]["IQR"]=descriptive_stats[column]["Q3:75%"]-descriptive_stats[column]["Q1:25%"]
        descriptive_stats[column]["1.5rule"]=1.5*descriptive_stats[column]["IQR"]
        descriptive_stats[column]["Lesser_outlier"]=descriptive_stats[column]["Q1:25%"]-descriptive_stats[column]["1.5rule"]
        descriptive_stats[column]["Greater_outlier"]=descriptive_stats[column]["Q3:75%"]+descriptive_stats[column]["1.5rule"]
        descriptive_stats[column]["Min"]=dataset[column].min()
        descriptive_stats[column]["Max"]=dataset[column].max()
        descriptive_stats[column]["kurtosis"]=dataset[column].kurtosis()
        descriptive_stats[column]["skew"]=dataset[column].skew()
        descriptive_stats[column]["Var"]=dataset[column].var()
        descriptive_stats[column]["Std"]=dataset[column].std()
    return descriptive_stats
```

```python
Univariate(dataset, quantitative_columns)
```
