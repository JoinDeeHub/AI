```python
import pandas as pd
```

```python
#load the dataset
dataset = pd.read_csv('insurance_pre.csv')

#dummy the categorical variables
dataset = pd.get_dummies(dataset, drop_first=True)
# Display the first few rows of the dataset
dataset
```

```python
# Define the features and target variable
independent = dataset.drop('charges', axis=1)  
#print(independent)
dependent = dataset['charges']
```

```python
# Import necessary libraries for model training and evaluation
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeRegressor
```

```python
# Initialize the DecisionTreeRegressor with GridSearchCV
# Create a GridSearchCV object with DecisionTreeRegressor
GridSearchCV = GridSearchCV(
    estimator=DecisionTreeRegressor(),
    param_grid={
        'criterion': ['squared_error', 'friedman_mse', 'absolute_error', 'poisson'],
        'splitter': ['best', 'random'],
        'max_features': ['auto', 'sqrt', 'log2']
    },
    refit=True,
    verbose=3,
    n_jobs=-1
)

# Fit the model using GridSearchCV
GridSearchCV.fit(independent, dependent)


# R2score value for best parameters
print(f"GridSearchCV Results R2score value for best parameters: {GridSearchCV.best_params_}")
```

```python
# Get the results from GridSearchCV
Results = GridSearchCV.cv_results_
# Convert the results to a DataFrame for better readability
Results_df = pd.DataFrame.from_dict(Results)
Results_df
```

```python
age_input = int(input("Enter the age: "))
bmi_input = float(input("Enter the BMI: "))
children_input = int(input("Enter the number of children: "))
sex_male_input = int(input("Enter 0 if you are male, 1 otherwise: "))
smoker_input = int(input("Enter 0 if you are a smoker, 1 otherwise: "))
```

```python
# Create a DataFrame for the input data
input_data = pd.DataFrame({
    'age': [age_input],
    'bmi': [bmi_input],
    'children': [children_input],
    'sex_male': [sex_male_input],
    'smoker_yes': [smoker_input]
})
```

```python
# Make predictions using the best model
predictions = GridSearchCV.predict(input_data)
print(f"Predicted insurance charges: {predictions}")
```
